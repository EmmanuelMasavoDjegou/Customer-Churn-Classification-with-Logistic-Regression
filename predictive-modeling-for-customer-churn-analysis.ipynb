# <div style="font-size:20px;text-align:center;color:black;border-bottom:5px #0026d6 solid;padding-bottom:3%"> UNDERSTANDING LOGISTIC REGRESSION </div>

**Logistic Regression** is a statistical method used for binary classification tasks, where the goal is to predict whether an input belongs to one of two classes, typically represented as 0 or 1. Unlike linear regression, which predicts a continuous value, logistic regression aims to produce a probability that an input belongs to a particular class. Here's a summary of the key points you've provided:

- **Objective of Logistic Regression:** Logistic regression transforms the output of linear regression into a range suitable for binary classification (0 or 1), addressing the issue of linear regression's unbounded output.

- **Non-Linear Relationship:** Logistic regression recognizes that the relationship between the independent variable(s) and the predictor variable is not necessarily linear, accommodating more complex patterns in the data.

- **Non-Normal Distribution of Errors:** Unlike linear regression, logistic regression doesn't assume that the errors are normally distributed. This makes it more suitable for situations where the assumptions of linear regression are violated.

- **Modeling Probabilities:** Logistic regression models the mean of the response variable (probability, denoted as p) in terms of an explanatory variable (x), using the odds ratio. This allows us to estimate the likelihood of an input belonging to a particular class.
    
    - **Sigmoid Function:** The sigmoid function transforms the linear combination of input features (denoted as z) into a value between 0 and 1, which represents a probability. 
    
    $$\sigma(z) = \frac{1}{1 + e^{-z}}$$
    
    - **Log-Odds (Logit Function):** The log-odds, also known as the logit function, is the natural logarithm of the odds ratio. It maps the probability of the positive class to the real number line.
    
    $$\text{logit}(p) = \ln\left(\frac{p}{1 - p}\right)$$
    
    - **Probability Prediction:** The probability p that an input belongs to the positive class is obtained by applying the sigmoid function to the linear combination of input features.
    
    $$p = \sigma(z) = \frac{1}{1 + e^{-\left(\beta_0 + \beta_1 x_1 + \beta_2 x_2 + \ldots + \beta_n x_n\right)}}$$
    
 - **Cost Function:** The cost function of logistic regression is designed to be optimized during the training process. It's derived from the sigmoid function and aims to measure the difference between predicted probabilities and actual class labels.
    
    - **Cost Function (Log-Likelihood):** The cost function, also known as the log-likelihood, measures the discrepancy between predicted probabilities and actual class labels.
    
    $$\text{Cost}(p, y) = -\left[y \cdot \ln(p) + (1 - y) \cdot \ln(1 - p)\right]$$
    
 - **Assumptions of Logistic Regression:**
 
 - **Linearity:** Unlike linear regression, logistic regression doesn't assume a linear relationship between independent variables and the target variable.

    - **Binary Outcome: **The target variable must be binary, having only two categories or classes.

    - **No Distribution Assumptions:** The independent variable's distribution doesn't need to be normal, linearly related, or have equal variance within each group.

    - **Mutually Exclusive and Exhaustive Categories:** The classes being predicted must be distinct and cover all possible outcomes.

<div class="alert alert-block alert-success"> We consider the Shill_Bidding_Dataset.csv dataset, which contains details regarding auctions done for various products on eBay.com. The target column, Class, provides information about the bidding behavior, 0 being normal and 1 being abnormal behavior. Abnormal behavior can be similar to malicious clicks or automatic bidding. 

Let's develop a machine learning model that can predict whether the bidding behavior in a particular auction is normal (0) or not (1). We will apply linear and logistic regression to predict the output and check which one of them is useful in this situation. </div>

```python
# Import necessary libraries
import pandas as pd              # For data manipulation and analysis
import numpy as np               # For numerical operations
from sklearn.model_selection import train_test_split  # For splitting the dataset into training and testing sets
from sklearn import linear_model  # For using the linear regression model
import matplotlib.pyplot as plt  # For data visualization
```

```python
# Import necessary libraries
import pandas as pd  # For data manipulation and analysis

# Read the CSV file using pandas
data = pd.read_csv("//kaggle/input/shill-bidding-dataset/Shill Bidding Dataset.csv")

# Display the first few rows of the DataFrame to inspect the data
data.head()
```

<div class="alert alert-block alert-success"> Next, we'll eliminate columns that don't contribute to the case study's relevance. Specifically, we'll exclude the Record_ID, Auction_ID, and Bidder_ID columns. These columns consist of unique identifiers and don't provide additional meaningful insights to the analysis. 

In other words, they don't offer any new information that would enhance the model's understanding of the data. </div>

```python
# Drop the specified columns (Record_ID, Auction_ID, Bidder_ID) from the DataFrame
data.drop(["Record_ID", "Auction_ID", "Bidder_ID"], axis=1, inplace=True)

# Display the first few rows of the DataFrame after dropping the columns
data.head()
```

<div class="alert alert-block alert-success"> Let's divide the data into training and testing sets using the following approach. We'll maintain the default proportion for test data, which is 30% of the entire dataset. <br><br>
Additionally, we'll use a random_state value of 1 to ensure reproducibility. This means that anyone else running the code will obtain similar outcomes. <br><br>
To address potential class imbalances, we'll perform stratified splitting. This ensures that we consider the distribution of classes in the target column while creating the training and testing sets. </div>

```python
# Extract features by dropping the "Class" column
X = data.drop("Class", axis=1)

# Extract the target variable ("Class")
y = data["Class"]

# Split the dataset into training and testing sets
# X_train: Features for training, X_test: Features for testing
# y_train: Target labels for training, y_test: Target labels for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
# Print the sizes of the training and testing datasets
print("Training dataset size: {}, Testing dataset size: {}".format(X_train.shape, X_test.shape))
```

<div class="alert alert-block alert-success"> Observe that the proportion of the training and testing dataset sizes is 7:3, aligning with our initial calculation of the test set size (30%) and train set size (70%). </div>

###  Linear Regression Model

```python
# Import the LinearRegression model from scikit-learn
linear = linear_model.LinearRegression()

# Fit the LinearRegression model using the training data
linear.fit(X_train, y_train)

# The output of the above code will be the model information, which includes various parameters.

# Predict on the first 10 test data points using the trained model
predictions = linear.predict(X_test)[:10]

# Display the predicted values for the first 10 test data points
print("Predicted values using linear regression:", predictions)
```

```python
# Obtain the actual target values for the first 10 test data points
actual_values = y_test[:10].values

# Display the actual target values for the first 10 test data points
print("Actual target values:", actual_values)
```

<div class="alert alert-block alert-success"> As evident, the linear regression model didn't predict discrete values of 1 and 0 as intended; instead, it produced continuous outputs, as we observed in the earlier step. This outcome signifies that the model isn't effectively discerning between the distinct classes. </div>

```python
# Evaluate the score of the linear regression model on the training and testing datasets

# Calculate and print the score on the training dataset
score_train = linear.score(X_train, y_train)
print("Score on training dataset using linear regression:", score_train)

# Calculate and print the score on the testing dataset
score_test = linear.score(X_test, y_test)
print("Score on testing dataset using linear regression:", score_test)
```

###  Logistic Regression Model

```python
# Fit the model using logistic regression
logit = linear_model.LogisticRegression()
logit.fit(X_train, y_train)

# Predict on the test data using the logistic regression model
logit_predictions = logit.predict(X_test)[:10]

# Display the predicted values obtained using logistic regression
print("Predicted values using logistic regression:", logit_predictions)
```

```python
# Display the actual target values for the first 10 test data points
actual_values = y_test[:10].values
print("Actual target values:", actual_values)
```

```python
# Calculate and print the score on the training dataset using logistic regression
logit_score_train = logit.score(X_train, y_train)
print("Score on training dataset using logistic regression:", logit_score_train)

# Calculate and print the score on the testing dataset using logistic regression
logit_score_test = logit.score(X_test, y_test)
print("Score on testing dataset using logistic regression:", logit_score_test)
```

###  Models Comparison

<div class="alert alert-block alert-success"> From this case study, we can draw a couple of noteworthy observations about the logistic regression algorithm. <br><br> 
Firstly, unlike linear regression that yielded continuous outputs, logistic regression is capable of predicting discrete classes as outcomes. 
Secondly, it's evident that logistic regression achieved a significantly higher score in comparison to linear regression. 
This disparity in scores indicates that logistic regression outperformed linear regression and is better suited for classification tasks. </div>

####  Model Building

####  Using Statsmodel Package

<div class="alert alert-block alert-success"> In this instance, the statsmodels package will be employed to fit a logistic regression model. This decision is due to the package's advantageous features, including the availability of a summary and parameter information, which were not present in earlier performed tasks. </div>

After acquiring the importance values of all features in the previous task, your current objective is to construct a logistic regression model utilizing the five most significant features to predict customer churn. 

The customer's attributes consist of the following: **Age: 50, EstimatedSalary: 100,000, CreditScore: 600, Balance: 100,000, NumOfProducts: 2**. 

Logistic regression is selected as the foundational model for churn prediction due to its straightforward interpretability.

```python
import statsmodels.api as sm

# Select the top 5 important features
top5_features = ['Age', 'EstimatedSalary', 'CreditScore', 'Balance', 'NumOfProducts']

# Build the logistic regression model
logReg = sm.Logit(y_train, X_train[top5_features])
logistic_regression = logReg.fit()
```

```python
# Print the summary of the logistic regression model
print(logistic_regression.summary())

# Print the parameters (coefficients) of the model
print(logistic_regression.params)
```

<div class="alert alert-block alert-success"> We can observe that the parameters for EstimatedSalary and Balance are relatively small, indicating that their impact on the target variable is much weaker compared to the other three features. </div>

```python
# Create a function to compute the predicted log-odds

coef = logistic_regression.params

def predicted_log_odds(coef, Age, EstimatedSalary, CreditScore, Balance, NumOfProducts):
    """
    Compute the predicted log-odds of churning based on logistic regression coefficients and feature values.
    
    Parameters:
    coef (array-like): Coefficients from the logistic regression model.
    Age (float): Customer's age.
    EstimatedSalary (float): Estimated salary of the customer.
    CreditScore (float): Credit score of the customer.
    Balance (float): Customer's bank balance.
    NumOfProducts (float): Number of products purchased by the customer.
    
    Returns:
    float: Predicted log-odds of churning.
    """
    return coef[0] * Age + coef[1] * EstimatedSalary + coef[2] * CreditScore + coef[3] * Balance + coef[4] * NumOfProducts
```

<div class="alert alert-block alert-success"> This function computes the predicted log-odds of churning based on the logistic regression coefficients and the values of the five selected features. </div>

```python
import numpy as np

# Calculate the predicted log-odds using the coefficients and feature values
predicted_log_odds = predicted_log_odds(coef, 50, 100000, 600, 100000, 2)

# Calculate the probability of churning using the predicted log-odds
probability_of_churning = np.exp(predicted_log_odds) / (1 + np.exp(predicted_log_odds))

# Print the probability
print("The customer has a {:.2f}% chance of churn.".format(probability_of_churning * 100))
```

####  Using Scikit-Learn Package

```python
from sklearn.linear_model import LogisticRegression

# Create a LogisticRegression model with specified parameters
clf = LogisticRegression(random_state=0, solver='lbfgs')

# Fit the model on the training data using the top 5 features
clf.fit(X_train[top5_features], y_train)

# Make predictions on the test data
predictions = clf.predict(X_test[top5_features])

# Get the predicted probabilities for each class
predicted_probabilities = clf.predict_proba(X_test[top5_features])

# Calculate the accuracy score of the model on the test data
accuracy = clf.score(X_test[top5_features], y_test)

# Print the predicted class labels, predicted probabilities, and accuracy score
print("Predicted Labels:", predictions)
```

```python
print("Predicted Probabilities:", predicted_probabilities)
```

```python
print("Accuracy Score:", accuracy)
```
